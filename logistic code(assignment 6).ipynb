# Bank problem statement

## importing librares

```python
import pandas as pd
```

importing data


```python
df=pd.read_csv('bank-full.csv',sep=';')
df
```

## Intial Analysis

```python
df.dtypes
```

```python
df.isna().sum()
```

## Data preparation

Replacing yes or no with 1 and 0

```python
df=df.replace({'default':{'yes':1,'no':0}})
df=df.replace({'housing':{'yes':1,'no':0}})
df=df.replace({'loan':{'yes':1,'no':0}})
df=df.replace({'y':{'yes':1,'no':0}})
df.head()
```

Replacing object data with dummy values

```python
df_objectdata=df.select_dtypes(include=['object'])
df_objectdata.head()
```

```python
dummies=pd.get_dummies(df_objectdata,drop_first=True)
dummies
```

```python
df_main=pd.concat([df,dummies],axis=1)
df_main.drop(columns=df_objectdata.columns,axis=1,inplace=True)
df_main
```

```python
X=df_main.drop(labels='y',axis=1)
y=df_main[['y']]
```

```python
X
```

```python
y
```

## Train test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.20,random_state=12)
```

## Model Training


```python
import warnings
warnings.filterwarnings('ignore')

from sklearn.linear_model import LogisticRegression
log_reg=LogisticRegression()
log_reg.fit(X,y)
```

## Model testing

Train data

```python
y_train_pred=log_reg.predict(X_train)
```

Test data

```python
y_test_pred=log_reg.predict(X_test)
```

# Model Evaluation

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score
```

For Training data

```python
accuracy_score(y_train,y_train_pred)
```

```python
confusion_matrix(y_train,y_train_pred)
```

```python
precision_score(y_train,y_train_pred)
```

```python
recall_score(y_train,y_train_pred)
```

For test data

```python
accuracy_score(y_test,y_test_pred)
```

```python
confusion_matrix(y_test,y_test_pred)
```

```python
precision_score(y_test,y_test_pred)
```

```python
recall_score(y_test,y_test_pred)
```
